```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load CSV file
file_path = "results.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Visualization 1: Heatmap of Sharpe Ratio
pivot_sharpe = data.pivot(index="ShortWindow", columns="LongWindow", values="SharpeRatio")
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_sharpe, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Sharpe Ratio Heatmap")
plt.xlabel("Long Window")
plt.ylabel("Short Window")
plt.show()

# Visualization 2: Heatmap of Max Drawdown
pivot_drawdown = data.pivot(index="ShortWindow", columns="LongWindow", values="MaxDrawdown")
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_drawdown, annot=True, fmt=".2f", cmap="YlGnBu", cbar=True)
plt.title("Max Drawdown Heatmap")
plt.xlabel("Long Window")
plt.ylabel("Short Window")
plt.show()

# Visualization 3: Final Capital vs Short and Long Window (3D Scatter)
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data["ShortWindow"], data["LongWindow"], data["FinalCapital"], c=data["FinalCapital"], cmap="viridis")
ax.set_title("Final Capital vs Short and Long Window")
ax.set_xlabel("Short Window")
ax.set_ylabel("Long Window")
ax.set_zlabel("Final Capital")
fig.colorbar(scatter, ax=ax, label="Final Capital")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "bollinger_bands_results.csv"
data = pd.read_csv(file_path)

# Display the first few rows to ensure data is loaded correctly
print(data.head())

# Visualization 1: Line Plot for Sharpe Ratio vs SMA Window
plt.figure(figsize=(10, 6))
plt.plot(data["SmaWindow"], data["SharpeRatio"], marker='o', label="Sharpe Ratio")
plt.title("Sharpe Ratio vs SMA Window")
plt.xlabel("SMA Window")
plt.ylabel("Sharpe Ratio")
plt.grid(True)
plt.legend()
plt.show()

# Visualization 2: Line Plot for Max Drawdown vs SMA Window
plt.figure(figsize=(10, 6))
plt.plot(data["SmaWindow"], data["MaxDrawdown"], marker='o', color='orange', label="Max Drawdown")
plt.title("Max Drawdown vs SMA Window")
plt.xlabel("SMA Window")
plt.ylabel("Max Drawdown")
plt.grid(True)
plt.legend()
plt.show()

# Visualization 3: Line Plot for Final Capital vs SMA Window
plt.figure(figsize=(10, 6))
plt.plot(data["SmaWindow"], data["FinalCapital"], marker='o', color='green', label="Final Capital")
plt.title("Final Capital vs SMA Window")
plt.xlabel("SMA Window")
plt.ylabel("Final Capital")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "z_score_results.csv"
data = pd.read_csv(file_path)

# Display the first few rows of data to confirm loading
print(data.head())

# Visualization 1: Sharpe Ratio vs Z-Score Window
plt.figure(figsize=(10, 6))
plt.plot(data["ZScoreWindow"], data["SharpeRatio"], marker='o', label="Sharpe Ratio", color="blue")
plt.title("Sharpe Ratio vs Z-Score Window")
plt.xlabel("Z-Score Window")
plt.ylabel("Sharpe Ratio")
plt.grid(True)
plt.legend()
plt.show()

# Visualization 2: Max Drawdown vs Z-Score Window
plt.figure(figsize=(10, 6))
plt.plot(data["ZScoreWindow"], data["MaxDrawdown"], marker='o', label="Max Drawdown", color="red")
plt.title("Max Drawdown vs Z-Score Window")
plt.xlabel("Z-Score Window")
plt.ylabel("Max Drawdown")
plt.grid(True)
plt.legend()
plt.show()

# Visualization 3: Final Capital vs Z-Score Window
plt.figure(figsize=(10, 6))
plt.plot(data["ZScoreWindow"], data["FinalCapital"], marker='o', label="Final Capital", color="green")
plt.title("Final Capital vs Z-Score Window")
plt.xlabel("Z-Score Window")
plt.ylabel("Final Capital")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the results from the CSV file
data = pd.read_csv("/Users/ojasjain/Desktop/Internships/LongShanks Capital/momentum_mean_reversion_results.csv")

# Display the first few rows of the data
print(data.head())

# Visualization 1: Heatmap for Sharpe Ratio
pivot_sharpe = data.pivot_table(
    index="RSIPeriod", 
    columns="BollingerWindow", 
    values="SharpeRatio", 
    aggfunc="mean"
)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_sharpe, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Sharpe Ratio Heatmap")
plt.xlabel("Bollinger Window")
plt.ylabel("RSI Period")
plt.show()

# Visualization 2: 3D Surface Plot for Final Capital
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

x = data["RSIPeriod"]
y = data["BollingerWindow"]
z = data["FinalCapital"]

scatter = ax.scatter(x, y, z, c=z, cmap="viridis", s=50)
ax.set_xlabel("RSI Period")
ax.set_ylabel("Bollinger Window")
ax.set_zlabel("Final Capital")
plt.colorbar(scatter, ax=ax, label="Final Capital")
plt.title("Final Capital Across RSI and Bollinger Window")
plt.show()

# Visualization 3: Line Plot for Max Drawdown
plt.figure(figsize=(12, 8))
sns.lineplot(data=data, x="MomentumPeriod", y="MaxDrawdown", hue="RSIPeriod", style="BollingerWindow", markers=True)
plt.title("Max Drawdown by Momentum Period")
plt.xlabel("Momentum Period")
plt.ylabel("Max Drawdown")
plt.legend(title="RSI Period and Bollinger Window")
plt.show()
```
